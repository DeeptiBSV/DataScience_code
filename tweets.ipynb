```python
import pandas as pd
tweets = pd.read_csv("tweets.csv", encoding = "ISO-8859-1")
tweets.head()
```

```python
tweets["text"].head()
```

```python
import nltk
nltk.download('stopwords')
```

```python
##preprocess the text data
import string
punctuation = string.punctuation
from nltk.corpus import stopwords
stopwords = stopwords.words("english")

def _clean(text):
    
    text= text.lower()
    text = "".join(x for x in text if x not in punctuation)
    
    words = text.split()
    words = [w for w in words if w not in stopwords ]
    text = " ".join(words)
    return text

_clean("I am a person!!")
```

```python
 tweets["cleaned"] = tweets["text"].apply(_clean)
```

```python
tweets[["cleaned","text"]]
```

```python
import collections
from collections import Counter
entire_text = " ".join(tweets["cleaned"])

entire_text = entire_text.split()
Counter(entire_text).most_common(50)
```

```python
raw_text = " ".join(tweets["text"])
mentions = [w for w in raw_text.split() if w.startswith("#")]
Counter(mentions).most_common(50)
#print(mentions)
```

```python
raw_text = " ".join(tweets["text"])
mentions = [w for w in raw_text.split() if w.startswith("@")]
Counter(mentions).most_common(50)
#print(mentions)
```

```python
raw_text = " ".join(tweets["text"])
mentions = [w for w in raw_text.split() if w.startswith("#")]
mentions = [w for w in mentions if "demo" not in w.lower()]
Counter(mentions).most_common(50)
```

```python
from nltk import ngrams
bigrams = ngrams(entire_text, 4)
Counter(bigrams).most_common(10)
```
